```python
# CHATBOT DE SUPORTE GOODWE

from google import genai
from google.genai import types


# CONFIGURAÇÃO DA API GEMINI


API_KEY = "AIzaSyAuhFo4swXAZPSObKodTjOU8xfE4L2jCPU"

client = genai.Client(api_key=API_KEY)

MODELO = "gemini-2.5-flash"


# DEFINIÇÃO DE PERSONAS
PERSONAS = {
    "1": {
        "nome": "Operador Comercial",
        "descricao": """
Você atende operadores comerciais da GoodWe.

Seu foco é:
- Explicar produtos
- Explicar benefícios dos inversores
- Explicar garantias
- Explicar monitoramento
- Explicar diferenciais técnicos de forma simples
- Ajudar vendedores e parceiros comerciais

Responda de forma objetiva, profissional e comercial.
"""
    },

    "2": {
        "nome": "Síndico",
        "descricao": """
Você atende síndicos de condomínios que usam ou querem usar soluções GoodWe.

Seu foco é:
- Explicar economia de energia
- Explicar funcionamento do sistema fotovoltaico
- Explicar segurança
- Explicar manutenção
- Explicar monitoramento da geração

Use linguagem simples, clara e sem excesso de termos técnicos.
"""
    },

    "3": {
        "nome": "Morador",
        "descricao": """
Você atende moradores residenciais que usam sistemas solares com GoodWe.

Seu foco é:
- Explicar geração solar
- Explicar aplicativo SEMS Portal
- Explicar economia
- Explicar dúvidas básicas
- Explicar possíveis alertas do sistema

Use linguagem amigável, simples e didática.
"""
    },

    "4": {
        "nome": "Técnico",
        "descricao": """
Você atende técnicos instaladores e profissionais de manutenção GoodWe.

Seu foco é:
- Diagnóstico de falhas
- Alarmes de inversores
- Configuração de equipamentos
- Comunicação Wi-Fi, LAN e datalogger
- Startup do sistema
- Manutenção técnica
- Verificação AC e DC

Use linguagem técnica, objetiva e detalhada.
"""
    }
}

# PERGUNTAS ESPERADAS + RESPOSTAS IDEAIS

FAQ = {
    "1": [
        {
            "pergunta": "Quais são os diferenciais dos inversores GoodWe?",
            "resposta": "Os inversores GoodWe oferecem alta eficiência, monitoramento remoto, proteções elétricas, boa compatibilidade com sistemas solares e excelente custo-benefício."
        },
        {
            "pergunta": "Como funciona o monitoramento?",
            "resposta": "O monitoramento pode ser feito pelo SEMS Portal, permitindo acompanhar geração, desempenho e alertas do sistema."
        },
        {
            "pergunta": "A GoodWe possui inversores híbridos?",
            "resposta": "Sim. A GoodWe possui modelos híbridos que podem trabalhar com baterias, dependendo do projeto e da compatibilidade do sistema."
        },
        {
            "pergunta": "Qual é a vantagem comercial da GoodWe?",
            "resposta": "A GoodWe combina tecnologia, confiabilidade, variedade de modelos e suporte para diferentes tipos de projetos solares."
        },
        {
            "pergunta": "Os inversores têm garantia?",
            "resposta": "Sim. A garantia depende do modelo e das condições comerciais aplicáveis. É recomendado verificar a documentação oficial do equipamento."
        }
    ],

    "2": [
        {
            "pergunta": "O condomínio pode economizar energia com sistema solar?",
            "resposta": "Sim. O sistema solar pode reduzir custos de áreas comuns, como iluminação, elevadores, bombas, portaria e garagem."
        },
        {
            "pergunta": "O sistema precisa de muita manutenção?",
            "resposta": "Não. Normalmente a manutenção é preventiva, com limpeza dos módulos e inspeções técnicas periódicas."
        },
        {
            "pergunta": "O sistema é seguro para o condomínio?",
            "resposta": "Sim, desde que instalado conforme normas técnicas e por profissionais qualificados."
        },
        {
            "pergunta": "Como o síndico acompanha a geração?",
            "resposta": "A geração pode ser acompanhada pelo SEMS Portal, que mostra dados de produção e desempenho do sistema."
        },
        {
            "pergunta": "Funciona em dias nublados?",
            "resposta": "Sim, mas a geração é menor porque há menos radiação solar disponível."
        }
    ],

    "3": [
        {
            "pergunta": "Como vejo a geração do meu sistema?",
            "resposta": "Você pode acompanhar a geração pelo aplicativo SEMS Portal, caso o sistema esteja configurado corretamente."
        },
        {
            "pergunta": "Por que minha geração caiu?",
            "resposta": "A queda pode ocorrer por clima nublado, sujeira nos módulos, sombreamento, falha de comunicação ou algum alerta no inversor."
        },
        {
            "pergunta": "O sistema gera energia à noite?",
            "resposta": "Não. Sistemas solares convencionais não geram energia à noite. Apenas sistemas com bateria podem usar energia armazenada."
        },
        {
            "pergunta": "Preciso desligar o inversor?",
            "resposta": "Normalmente não. O inversor opera automaticamente e só deve ser desligado em caso de manutenção ou orientação técnica."
        },
        {
            "pergunta": "O aplicativo está sem dados. O que fazer?",
            "resposta": "Verifique a internet, o roteador, o datalogger e se o inversor está conectado corretamente ao SEMS Portal."
        }
    ],

    "4": [
        {
            "pergunta": "O inversor está sem comunicação.",
            "resposta": "Verifique conexão Wi-Fi ou LAN, sinal do roteador, datalogger, antena, LEDs de comunicação e cadastro no SEMS Portal."
        },
        {
            "pergunta": "Apareceu erro de isolação.",
            "resposta": "Verifique cabos DC, conectores, módulos, umidade, string box e resistência de isolamento."
        },
        {
            "pergunta": "O inversor não inicia geração.",
            "resposta": "Verifique tensão DC mínima, polaridade das strings, disjuntores, seccionadoras, rede AC e mensagens de alarme."
        },
        {
            "pergunta": "O SEMS Portal não recebe dados.",
            "resposta": "Confirme internet ativa, datalogger online, número de série correto, sinal Wi-Fi e status de comunicação."
        },
        {
            "pergunta": "Como verificar falha de rede AC?",
            "resposta": "Meça tensão e frequência da rede, verifique disjuntores, conexões AC, configuração de país/rede e alarmes registrados."
        }
    ]
}

# FUNÇÕES DE INTERFACE

def mostrar_menu():
    print("=" * 60)
    print("CHATBOT DE SUPORTE GOODWE COM IA - GEMINI")
    print("=" * 60)

    print("\nEscolha a persona do atendimento:\n")
    print("1 - Operador Comercial")
    print("2 - Síndico")
    print("3 - Morador")
    print("4 - Técnico")


def mostrar_faq(opcao):
    print("\n" + "=" * 60)
    print("5 PERGUNTAS ESPERADAS COM RESPOSTAS IDEAIS")
    print("=" * 60)

    for i, item in enumerate(FAQ[opcao], start=1):
        print(f"\n{i}. Pergunta:")
        print(item["pergunta"])

        print("\nResposta ideal:")
        print(item["resposta"])

        print("-" * 60)


def criar_instrucao_sistema(persona):
    return f"""
Você é um chatbot de suporte da GoodWe.

Persona atendida:
{persona["nome"]}

Escopo do atendimento:
{persona["descricao"]}

Regras obrigatórias:
- Responda sempre em português do Brasil
- Seja educado, claro e objetivo
- Não invente dados técnicos específicos
- Quando não tiver certeza, recomende consultar o manual oficial ou o suporte técnico autorizado
- Para dúvidas de segurança elétrica, oriente que um profissional qualificado faça a verificação
- Mantenha o atendimento dentro do contexto de soluções GoodWe, inversores solares, monitoramento, instalação e suporte
"""


# FUNÇÃO PARA CONSULTAR GEMINI

def perguntar_ao_gemini(historico, pergunta_usuario):
    historico.append(f"Usuário: {pergunta_usuario}")

    conversa_formatada = "\n".join(historico)

    resposta = client.models.generate_content(
        model=MODELO,
        contents=conversa_formatada,
        config=types.GenerateContentConfig(
            temperature=0.4,
            max_output_tokens=1000
        )
    )

    texto = resposta.text

    historico.append(f"Assistente GoodWe: {texto}")

    return texto


# CHATBOT INTERATIVO

def iniciar_chat(persona):
    instrucao_sistema = criar_instrucao_sistema(persona)

    historico = [
        instrucao_sistema
    ]

    print("\n" + "=" * 60)
    print("CHATBOT INICIADO")
    print("Digite 'sair' para encerrar.")
    print("=" * 60)

    while True:
        pergunta = input("\nVocê: ").strip()

        if pergunta.lower() == "sair":
            print("\nGoodWe IA: Atendimento encerrado. Obrigado pelo contato!")
            break

        if pergunta == "":
            print("\nGoodWe IA: Digite uma pergunta ou 'sair' para encerrar.")
            continue

        try:
            resposta = perguntar_ao_gemini(historico, pergunta)
            print("\nGoodWe IA:")
            print(resposta)

        except Exception as erro:
            print("\nErro ao consultar o Gemini:")
            print(erro)


# PROGRAMA PRINCIPAL

def main():
    mostrar_menu()

    opcao = input("\nDigite a opção desejada: ").strip()

    if opcao not in PERSONAS:
        print("\nOpção inválida. Execute novamente e escolha uma opção de 1 a 4.")
        return

    persona = PERSONAS[opcao]

    print("\nPersona selecionada:")
    print(persona["nome"])

    mostrar_faq(opcao)

    iniciar_chat(persona)



# EXECUÇÃO


if __name__ == "__main__":
    main()
```
