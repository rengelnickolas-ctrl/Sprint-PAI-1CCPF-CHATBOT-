```python
import google.generativeai as genai

API_KEY = "COLOQUE_SUA_API_KEY_AQUI" # OBS: O USUARIO DEVE CRIAR E COLOCAR SUA PROPRIA API KEY NO GEMINI

genai.configure(api_key=API_KEY)

MODELO_GEMINI = "gemini-2.5-flash"

# =========================
# DEFINIÇÃO DO ESCOPO
# =========================

personas = {
    "1": "operador comercial",
    "2": "síndico",
    "3": "morador",
    "4": "técnico"
}

print("Escolha a persona atendida pelo chatbot:")
print("1 - Operador comercial")
print("2 - Síndico")
print("3 - Morador")
print("4 - Técnico")

opcao = input("Digite o número da persona: ")

persona = personas.get(opcao, "técnico")

perguntas_esperadas = [
    "Como funciona a gestão de recarga para veículos elétricos?",
    "É possível controlar a demanda de energia em tempo real?",
    "Como pode ser feito o rateio de custos em condomínios?",
    "Existe integração com SEMS+ ou aplicativo de monitoramento?",
    "Como funciona o modelo comercial e de cobrança?"
]

contexto_injetado = """
Você é um chatbot de suporte da GoodWe focado em soluções de recarga para veículos elétricos.

Contexto:
- A solução envolve carregadores para veículos elétricos.
- O objetivo é transformar sessões de recarga em dados estruturados e inteligência acionável.
- Existem aplicações para setor comercial, varejo, condomínios e setor corporativo.
- Em áreas comerciais, o foco pode envolver otimização em tempo real, controle de demanda e tarifação dinâmica.
- Em condomínios, o foco pode envolver gestão compartilhada, rateio inteligente, síndico virtual e boa experiência do usuário.
- O equipamento de referência é da linha HCA G2.
- A interação e configuração podem ocorrer por aplicativos como SolarGo e SEMS+.
- O carregador possui luzes indicativas para sinalização de status.
- No momento, o suporte via API para EV Chargers pode não estar disponível diretamente.
- Quando houver API, o modelo de consulta será do tipo pull, ou seja, consultas manuais ou programadas.
- A solução deve considerar cobrança, responsáveis pelos custos de energia, gateway de pagamento, autenticação, divisão de receita e modelo de implantação.

Responda sempre de forma clara, objetiva e adequada à persona selecionada.
Caso a pergunta ou o contexto não tenha nada a ver com a GoodWe e seus produtos, responda de forma educada que não tem como ajudar e que não há informações sobre o assunto
Não mencione que as informações vieram de um slide ou apresentação.
"""

SYSTEM_PROMPT = f"""
Persona atendida: {persona}

Perguntas que o chatbot deve responder:
{chr(10).join("- " + pergunta for pergunta in perguntas_esperadas)}

Contexto injetado no modelo:
{contexto_injetado}
"""

# =========================
# CHATBOT INTERATIVO
# =========================

def chatbot_interativo():
    modelo = genai.GenerativeModel(
        model_name=MODELO_GEMINI,
        system_instruction=SYSTEM_PROMPT
    )

    chat = modelo.start_chat(history=[])

    print("=" * 60)
    print("💬 CHATBOT GoodWe⚡🔋")
    print(f"Persona selecionada: {persona}")
    print("Digite sua mensagem e pressione Enter.")
    print("Para encerrar, digite: sair")
    print("=" * 60)

    while True:
        entrada = input("\n👤 Você: ").strip()

        if entrada.lower() == "sair":
            print("\n🤖 Assistente: Obrigado por entrar em contato! Até logo! 👋")
            break

        if not entrada:
            print("⚠️ Digite algo ou 'sair' para encerrar.")
            continue

        resposta = chat.send_message(entrada)

        print("\n🤖 Assistente:")
        print(resposta.text)

chatbot_interativo()
```
